### 1. Move ECG of healty people to another location


#### Getting initial's data, CSV's data and SR positions


```python
def read_file(file_path: str) -> list:
    content = []

    with open(file_path, 'r') as lines:
        for line in lines:
            content.append(line.strip())

    return content

def find_hearth_disease_position(initials: list, hearth_disease: str) -> list:
    hearth_diseases_pos : list = []

    for index, initial in enumerate(initials, 0):
        if initial == hearth_disease:
            hearth_diseases_pos.append(index)

    return hearth_diseases_pos

def samples_filter(ecg_names: list, block_names: list) -> list:
    block_set = set(block_names)

    return [ecg for ecg in ecg_names if ecg not in block_set]
```

#### Moving healty ECG samples to another folder


```python
import os
import shutil

def move_ecg_samples(ecg_samples: list, heart_disease_positions: list, folders: list):
    origin_folder = folders[0]
    destiny_folder = folders[1]

    files_moved: int = 0

    if not os.path.exists(destiny_folder):
        os.makedirs(destiny_folder)

    for index, csv_name in enumerate(ecg_samples, 0):
        csv_name += '.csv'

        for hd_pos in heart_disease_positions:
            if hd_pos == index:
                origin_path = os.path.join(origin_folder, csv_name)
                destiny_path = os.path.join(destiny_folder, csv_name)

                if os.path.exists(origin_path):
                    try:
                        shutil.move(origin_path, destiny_path)
                        files_moved += 1
                        print(f"File {csv_name}.csv moved from {origin_folder} to {destiny_folder}")
                    except Exception as ex:
                        print(f"Error at move {csv_name}: {ex}")

    print(f"{files_moved} files moved to {destiny_folder}")
```

### Main


```python
if __name__ == "__main__":
    list_path: str = './ECG_Samples/list/'

    dat_names: list = ['diagnostics.dat', 'block_list.dat', 'filename.dat']

    ecg_origin_folder_path: str = './ECG_Samples/'
    folder_name: str = './ECG_BY_HEARTH_DISEASES/'
    folders: list = [ecg_origin_folder_path, folder_name]

    initials: list = read_file(list_path + dat_names[0])
    block_list: list = read_file(list_path + dat_names[1])
    ecg_names: list = read_file(list_path + dat_names[2])

    initials_set = set(initials)

    for init in initials_set:
        heart_disease_positions: list = find_hearth_disease_position(initials, init)
        clean_ecg_names: list = samples_filter(ecg_names, block_list)
        folders[1] += init
        move_ecg_samples(clean_ecg_names, heart_disease_positions, folders)
        folders[1] = folder_name
```
